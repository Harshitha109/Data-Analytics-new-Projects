Importing Libraries

```python
import pandas as pd 
import numpy as np
```

```python
file_name = "HR_Analytics.csv"
dataset = pd.read_csv(file_name)
dataset
```

```python
print(dataset.info())
dataset.isnull().sum()
```

```python
# Count null values in each column
dataset['YearsWithCurrManager'] = dataset['YearsWithCurrManager'].fillna(0)
print(dataset.isnull().sum())
#print("Null values check after")
#dataset.isnull().sum()
```

```python
duplicates = dataset.duplicated().sum()
print(f"Total duplicate rows: {duplicates}")

cleaned_dataset = dataset.drop_duplicates()
print(f"Total duplicate rows: {cleaned_dataset.duplicated().sum()}")
```

```python
# Converting identifier columns to string type to ensure they are treated as categorical labels rather than numeric values
# EmpID and EmployeeNumber are unique identifiers, so they should be stored as strings to avoid any mathematical operations
```

```python
cleaned_dataset['EmpID'] = cleaned_dataset['EmpID'].astype(str)
cleaned_dataset['EmployeeNumber'] = cleaned_dataset['EmployeeNumber'].astype(str)

# Converting YearsWithCurrManager to integer type as it represents whole years of tenure with the current manager
# This ensures consistent data type for numerical analysis and modeling
cleaned_dataset['YearsWithCurrManager'] = cleaned_dataset['YearsWithCurrManager'].astype(int)

# Defining list of categorical columns that should be treated as categories in statistical analysis
# These columns contain discrete, finite set of values representing different groups
categorical_cols = [
    'Attrition', 'Gender', 'OverTime', 'BusinessTravel', 'Department',
    'EducationField', 'JobRole', 'MaritalStatus', 'AgeGroup', 'SalarySlab'
]
# Converting each categorical column to category data type to optimize memory usage
for col in categorical_cols:
    cleaned_dataset[col] = cleaned_dataset[col].astype('category')

# Displaying data types of all columns to verify the conversion is correct
print(cleaned_dataset.dtypes)
print(cleaned_dataset.shape)

# Saving the cleaned and type-converted dataset to a CSV file for visualization
# The index=False parameter prevents pandas from adding an extra index column to the file
cleaned_dataset.to_csv('HR_Analytics_Cleaned.csv', index=False)

#A print message to ensure that the file is saved
print("Cleaned dataset saved as HR_Analytics_Cleaned.csv")
```
